```python
import kagglehub
import pandas as pd

path = kagglehub.dataset_download("tmdb/tmdb-movie-metadata")
df = pd.read_csv(path + "/tmdb_5000_movies.csv")

df.info()
```

```python
df.head()
```

```python
df['genres'].unique()
```

```python
import json

def extract_genres(genre_str):
    try:
        genres = json.loads(genre_str)  # transforma a string em lista de dicionários
        return [g["name"] for g in genres]  # pega só o campo "name"
    except:
        return []

df["genre_names"] = df["genres"].apply(extract_genres)

df.info()
```

```python
# "desempacotar" as listas e pegar valores únicos
unique_genres = set(genre for sublist in df['genre_names'] for genre in sublist)
print(unique_genres)
```

```python
# Média, mediana, desvio padrão e variância do Orçamento
print('Média: ', df['budget'].mean())
print('Mediana: ', df['budget'].median())
print('Desvio padrão: ', df['budget'].std())
print('Variância: ', df['budget'].var())
```

```python
# Média, mediana, desvio padrão e variância da Receita
print(f'Média: {df["revenue"].mean():.2f}')
print(f'Mediana: {df["revenue"].median():.2f}')
print(f'Desvio padrão: {df["revenue"].std():.2f}')
print(f'Variância: {df["revenue"].var():.2f}')
```

```python
# Média, mediana, desvio padrão e variância da Nota Média (vote_average)
print(f'Média: {df["vote_average"].mean():.2f}')
print(f'Mediana: {df["vote_average"].median():.2f}')
print(f'Desvio padrão: {df["vote_average"].std():.2f}')
print(f'Variância: {df["vote_average"].var():.2f}')
```

```python
# Média, mediana, desvio padrão e variância da Popularidade (popularity)
print(f'Média: {df["popularity"].mean():.2f}')
print(f'Mediana: {df["popularity"].median():.2f}')
print(f'Desvio padrão: {df["popularity"].std():.2f}')
print(f'Variância: {df["popularity"].var():.2f}')
```

### 1. Budget (Orçamento):
 - A média (29045039.88) é significativamente maior que a mediana (15000000.0). Isso sugere que existem alguns filmes com orçamentos muito altos puxando a média para cima. A distribuição dos orçamentos provavelmente é assimétrica à direita.
 - O desvio padrão (40722391.26) é alto em relação à média, indicando uma grande dispersão nos valores do orçamento.
 - A variância (1.66e+15) é um número muito grande, o que também reforça a ideia de uma grande variação nos orçamentos dos filmes.

### 2. Revenue (Receita):
 - Assim como o orçamento, a receita também apresenta uma média (82260638.65) muito maior que a mediana (19170001.00). Isso indica que poucos filmes de grande sucesso geram a maior parte da receita, criando uma distribuição altamente assimétrica à direita.
 - O desvio padrão (162857100.94) e a variância (2.65e+16) são extremamente altos, confirmando a enorme dispersão e a presença de outliers (filmes com receitas muito acima da média).

### 3. Vote Average (Nota Média):
 - A média (6.09) e a mediana (6.20) são bem próximas. Isso sugere que a distribuição das notas médias é mais simétrica do que as de orçamento e receita, com a maioria dos filmes tendo notas em torno de 6.
 - O desvio padrão (1.19) é relativamente baixo, indicando que a maioria das notas está concentrada perto da média.
 - A variância (1.43) também é baixa, confirmando a menor dispersão das notas em comparação com orçamento e receita.

### 4. Popularity (Popularidade):
 - A média (21.49) é consideravelmente maior que a mediana (12.92). Isso indica que alguns filmes são significativamente mais populares do que a maioria, gerando uma distribuição assimétrica à direita, embora não tão acentuada quanto orçamento e receita.
 - O desvio padrão (31.82) é alto em relação à mediana, mostrando uma dispersão notável na popularidade dos filmes.
 - A variância (1012.30) reflete essa dispersão.

### Concluindo:
As distribuições de orçamento e receita são fortemente influenciadas por filmes de muito alto orçamento e grande sucesso de bilheteria (outliers), resultando em médias muito superiores às medianas e alta variabilidade.

A distribuição da nota média é mais "normal" ou simétrica, com a maioria dos filmes recebendo notas em torno de 6 e menor variabilidade.

A popularidade também apresenta alguma assimetria à direita, indicando que poucos filmes são extremamente populares, mas a dispersão não é tão extrema quanto a de orçamento e receita.

Essa análise inicial sugere que, ao trabalhar com orçamento, receita e popularidade, pode ser interessante considerar o uso da mediana como uma medida de tendência central mais representativa, já que a média é muito afetada pelos valores extremos. Também pode ser útil explorar visualizações (como histogramas e box plots) para entender melhor a forma dessas distribuições e identificar outliers.

```python
# Histograma de Distribuição de Nota Média (vote_average) matplotlib
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(df['vote_average'], bins=20, edgecolor='black')
plt.title('Distribuição de Nota Média')
plt.xlabel('Nota Média')
plt.ylabel('Frequência')
plt.show()
```

```python
# Histograma de Distribuição de Nota Média (vote_average) NO seaborn:
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['vote_average'], bins=20, kde=True)
plt.title('Distribuição de Nota Média')
```

```python
# Gráficos de dispersão (ex: orçamento vs. receita)
plt.figure(figsize=(10, 6))
plt.scatter(df['budget'], df['revenue'])
plt.title('Orçamento vs. Receita')
plt.xlabel('Orçamento')
plt.ylabel('Receita')
```

```python
# Gráficos de dispersão (ex: orçamento vs. receita) no seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='budget', y='revenue')
plt.title('Orçamento vs. Receita')
plt.xlabel('Orçamento')
plt.ylabel('Receita')
```

```python
# Boxplot (ex: avaliação por gênero) no matplotlib
plt.figure(figsize=(10, 6))
plt.boxplot(df['vote_average'])
plt.title('Avaliação por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Nota Média')
plt.show()
```
